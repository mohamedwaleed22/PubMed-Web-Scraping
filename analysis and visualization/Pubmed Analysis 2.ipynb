# Top 10 citated (Percentage)

```python


sum_by_country = merged_df.groupby('Country')['N_Citations'].sum()  # Calculate the sum instead of the mean
total_citations = sum_by_country.sum()  # Calculate the total citations

sum_by_country = sum_by_country.sort_values(ascending=False)
top_10_countries = sum_by_country.head(10)

percentages = (top_10_countries / total_citations) * 100  # Calculate the percentage

plt.figure(figsize=(12, 7))
bar_plot = plt.bar(top_10_countries.index, percentages)  # Use the percentages instead of the mean values
plt.title('Top 10 Countries in Percentage of Total Citations')  # Update the title
plt.xticks(rotation=45)
plt.tight_layout()

# Adjust the text label positions and alignment
texts = []
for rect in bar_plot:
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x() + width / 2
    y = height + 0.5  # Adjust the vertical position
    alignment = 'center'  # Set the text alignment
    texts.append(plt.text(x, y, f"{height:.2f}%", ha=alignment, va='bottom'))  # Display the percentage with two decimal places

# Apply adjustments to avoid text overlap
adjust_text(texts, add_objects=bar_plot, autoalign='xy', only_move={'points':'y', 'text':'y'})

plt.show()
# plt.savefig('D:/projects/medical_research2/results/res/toptenCountriesbyPercentage.png')
```
